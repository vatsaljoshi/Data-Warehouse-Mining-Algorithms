```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn_extra.cluster import KMedoids
from sklearn.datasets import make_blobs
```

```python
Centers=[[2,2],[-2,2],[-2,-2]]
X, labels_true=make_blobs(n_samples=350,centers=Centers,cluster_std=0.4,random_state=0)
```

```python
plt.plot(X[:,0],X[:,1],'o',markerfacecolor='green',markersize=5)
```

```python
#from google.colab import files
#data_to_load = files.upload()
```

```python
kmedoids_model=KMedoids(n_clusters=3).fit(X)
labels= kmedoids_model.labels_
unique_labels=set(labels)
colors=[plt.cm.Spectral(each) for each in np.linspace
        (0,1,len(unique_labels))]

for k, col in zip(unique_labels,colors):
  class_member_mask=labels==k
  xy=X[class_member_mask]
  plt.plot(
      xy[:,0],
      xy[:,1],
      "o",
      markerfacecolor=tuple(col),
      markeredgecolor="k",
      markersize=8
  )
  plt.plot(
      kmedoids_model.cluster_centers_[:,0],
      kmedoids_model.cluster_centers_[:,1],
      ".",
      markerfacecolor='aqua',
      markeredgecolor='black',
      markersize=22

  )
  plt.title("k-medoid clustering \n medoids are marked as star")
```
